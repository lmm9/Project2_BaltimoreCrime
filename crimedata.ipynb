```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np 
from datetime import date
```

```python
#Importing election data 
csv_file = "BPD_Part_1_Victim_Based_Crime_Data.csv"
# Read our Data file with the pandas library
crimedata = pd.read_csv(csv_file, encoding="utf-8")
# Show just the header
crimedata.head()
```

```python
crimedata.dtypes
```

```python
crimedata['CrimeDate']= pd.to_datetime(crimedata['CrimeDate'], format='%m/%d/%Y')
crimedata.head()
```

```python
#Filter out crime with description = homicide 
homicide_data = crimedata.loc[crimedata['Description'] == "HOMICIDE"]
homicide_data.head()
```

```python
#Filter out crime date in 2020
homicide2020_data = homicide_data.loc[homicide_data['CrimeDate'] > "12/31/2019"]
homicide2020_data
```

```python
homicide2020_data=homicide2020_data.reset_index()
homicide2020_data.head()
```

```python
homicide2020_data['id'] = homicide2020_data.index
homicide2020_data.head()
```

```python
del homicide2020_data['index']
```

```python
homicide2020_data.set_index('id', inplace = True)
```

```python
homicide2020_data
```

```python
rds_connection_string = "postgres:Blue14horse@localhost:5432/balt_crime"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
conn = engine.connect()
```

```python
engine.table_names()
```

```python
homicide2020_data.to_sql(name = 'crime_data', con = engine, if_exists = 'append', index = True)
```

```python
crimedata_final = pd.read_sql('select * from crime_data' ,conn)
```

```python
# Export file as a CSV, without the Pandas index, but with the header
crimedata_final .to_csv("crimedata_final.csv", index=False, header=True)
```
